```python
import numpy as np
```

```python
#array
a = np.array([1, 2, 3, 4, 5])
print(a)
```

```python
M = np.array([[1, 2, 3],
              [4, 5, 6]])
print(M)
print(M.shape) #(2,3) - 2 raws, 3 columns
```

```python
#vectorized operations
a = np.array([1, 2, 3, 4])
print(a + 10)
print(a * 2)
print(a ** 2) #powered
```

```python
#random numbers
np.random.seed(42) #фіксує випадковість
r = np.random.random(5) # 5 чисел від 0 до 1
z = np.random.standard_normal(5) # 5 чисел із нормального стандартного розподілу
print(z) 
```

```python
#broadcasting
M = np.array([[1,2,3],
               [4, 5, 6]])
print(M + 100)
row = np.array([10, 20, 30])
print(M + row)
```

```python
#slicing
a = np.array([10,20,30,40,50])
print(a[0])
print(a[-1])
print(a[1:4])
print(a[a > 25])
```
