```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
data = yf.download("AAPL", period="1y")
```

```python
data.head()
```

```python
data["Close"].plot(figsize=(12, 5), title="AAPL Closing Price - Last 1 Year")
plt.ylabel("Price ($)")
plt.xlabel("Date")
plt.grid(True)
plt.show()
```
