```python
import numpy as np 
import yfinance as yf
import matplotlib.pyplot as plt

data = yf.download("AAPL", period="1y")
```

```python
close = data["Close"]
log_returns =np.log(close / close.shift(1))
log_returns.head()
```

```python
log_returns = log_returns.dropna() #removes NaN

log_returns.plot(kind="hist", bins=50,figsize=(10, 5), #histogram, 50 bars
                 title="AAPL Daily Log_Returns - Distribution")
plt.xlabel("Daily log-return")
plt.show()
```

```python
mu = log_returns.mean() #average daily log-return
sigma = log_returns.std() #spread of the bell, volatility (daily)
print("Daily drift (mu):        ", mu)
print("Daily volatility (sigma):", sigma)
```

```python
trading_days = 252
annual_mu = mu * trading_days
annual_sigma = sigma * np.sqrt(trading_days)

print("Annual drift:     ", annual_mu)
print("Annual volatility:", annual_sigma)
```
